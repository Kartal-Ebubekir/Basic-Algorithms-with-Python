## Asimptotik Notasyonlar


### Big Oh Notasyonu O(n)
- Zaman karmaşıklığında üst sınırı gösterir.  
- Çalışma süresi (n), algoritmanın çalıştıgı veri kümesideki değer sayısı kabul edilir.  
- O(n) seklinde gösterilir.    

| **Zaman karmasıklıgı T(n)** | **Algoritma** | **Açıkklama** |
|----------------------|---------------|------------------|
| **O(1)** | Direkt elemana ulasım. | Sabit zamanlı işlemdir çünkü dizideki herhangi bir elemana indeksle erişmek, dizinin boyutundan bağımsızdır.|
| **O(n)** | Linear search |Her eleman tek tek kontrol edilir, bu yüzden işlem sayısı n ile orantılıdır. |
| **O(n²)** | Bubble Sort, Selection Sort, Insertion Sort | İç içe döngüler nedeniyle işlem sayısı n² ile orantılıdır. |
| **O(n log n)** | Quicksort | Ortalama durumda böl-ve-yönet yaklaşımıyla çalışır; en kötü durumda O(n²) olabilir. |  



[![](https://www.w3schools.com/dsa/img_plot_timecomplexity.png)](https://www.w3schools.com/dsa/img_plot_timecomplexity.png)  



#### O(n) Nasıl hesaplanır ?  
- Sabit zamanlı işlemler: Atamalar, basit aritmetik işlemler, karşılaştırmalar, diziye erişim (O(1)) 1 kez sayılır.  
- Döngü veya Özyineleme: Girdinin boyutuyla (n) doğrudan orantılı olarak tekrar eden işlemler n cinsinden ifade edilir.  
- Eğer bir döngü, girdi boyutuna (n) bağlı olarak tam n kez çalışıyorsa, bu döngü O(n) karmaşıklığına neden olur.  
- T(n) fonksiyonunu O(n)'e indirgemek için sadece en hızlı büyüyen terim korunur.

| Kural | Örnek Fonksiyon (T(n)) | Büyük O İfadesi | Açıklama |
| :--- | :--- | :--- | :--- |
| Sabit katsayıyı atma | T(n) = 3n | O(n) | 3 gibi sabit çarpanlar atılır. n çok büyüdüğünde 3'ün etkisi ihmal edilir. |
| Sabit terimleri atma | T(n) = 5n + 100 | O(n) | 100 gibi sabitler veya n'den daha yavaş büyüyen terimler atılır. |

### Uzaysal Karmaşıklık S(n)

- S(n), bir algoritmanın çalışması sırasında ihtiyaç duyduğu toplam bellek (hafıza) miktarını girdi boyutuna (n) bağlı olarak ifade eden fonksiyondur.
- Örnek :  
```
def create_new_array(input_array):

    new_array = [0] * len(input_array)  
    
    for i in range(len(input_array)):
        new_array[i] = input_array[i]
        
    return new_array
```

- new_array : Fonksiyon, input_array ile aynı boyutta, yani n elemanlı yeni bir dizi (new_array) oluşturur.  

- Diğer değişkenler: i gibi diğer değişkenler için kullanılan bellek miktarı sabit ve giriş boyutundan bağımsızdır.
* Sonuç : Toplam bellek kullanımı, esas olarak oluşturulan yeni dizinin boyutuna bağlıdır.Yani, girdi büyüdükçe bellek kullanımı da aynı oranda artar.  
$$
S(n)=n+c
$$  
Burada c, sabit miktarda kullanılan bellek (örneğin sayaç değişkenleri) anlamına gelir.
Girdi boyutu büyüdükçe n terimi baskın olduğu için:  
$$
S(n) = n
$$  
şeklinde ifade edilir.

### Binary Search  
- Sırasıyla her eleman kontrol edilir.İstenen elemana ulasıldıgında kontrol islemi durdurulur.  
- Sadece sıralı listelerde kullanılabilir.Yani listenin küçükten büyüğe sıralı olması gerekir.  
- Her adımda ortanca eleman kontrol edilir.  
  - Eğer aranan değer ortanca elemandan küçükse, arama listenin sol yarısında devam eder.  
  - Eğer aranan değer ortanca elemandan büyükse, arama listenin sağ yarısında devam eder.  
  - Bu işlem, eleman bulunana kadar veya aralık boşalana kadar sürer.
- Arama alanı her adımda yarıya indirilir bu yüzden O(log n) karmaşıklığa sahiptir.  
-Zaman Karmaşıklığı: O(log n)

````python
def binary_search(arr, value):
  low = 0 # İslem yaptıgımız dizinin alt sınırı
  high = len(arr)-1 # İslem yaptığımız dizinin üst sınırı
  while low <= high:
    # median ortanca indeksi temsil ediyor.
    median = (low + high) // 2 # taşma riskini önlemek amacıyla low + (high-low) // 2  de kullanılabilir . Aynı sonucu verir ancak kullandığımız formül int'in maksimum kapasitesini (C,C++,Java gibi dillerde) aşabilir.

    if value == arr[median]:
      return median
    elif value < arr[median]:
      high = median - 1
    else :
      low = median  +1

  return -1 # Eğer aranan eleman listede yoksa
````

### Linear Search  
- Sırasıyla her eleman kontrol edilir.İstenen elemana ulasıldıgında kontrol islemi durdurulur.
- Karmasıklıgı O(n)  


````python
def linear_search(arr,value) -> int:
  for i in range(len(arr)):
    if arr[i] == value :
      return i
  return -1
````

````python
arr = [3, 7, 2, 9, 5]
value = 9
result = linear_search(arr, value)
````

### Bubble Sort  
- Her adımda bir eleman dizinin sonuna gider (İlk edımda en büyük eleman listenin sonuna gider , ikinci adımdaysa listenin en büyük ikinci elemanı sondan ikinci sıraya gider ve liste sıralanana kadar devam eder).  
- Zaman Karmaşıklığı: O(n²)

````python
def bubble_sort(arr) -> None:
  n = len(arr)
  for i in range(n-1): # n elemanlı bir liste için n-1 tur yeterlidir.Örnek : 2 elemanlı bir listede ilk turda büyük eleman listenin sonuna konur ve liste sıralanmış olur 2. tura gerek yok.
    for j in range(n-i-1): # Her turda en büyük elema sona gittiği için sondaki i kadar elemanı tekrar kontrol etmeye gerek yoktur. -1 ise j+1 yaparken "index out of range" hatası almamak içindir.
      if arr[j] > arr[j+1] :
        # swap
        arr[j],arr[j+1] = arr[j+1],arr[j]
````

### Selection Sort  
- Her adımda en kücük eleman listenin basına gider (İlk edımda en küçük eleman listenin başına gider , ikinci adımdaysa listenin en küçük ikinci elemanı ikinci sıraya gider ve liste sıralanana kadar devam eder).  
- Zaman Karmaşıklığı: O(n²)  
[![](https://mollify.noroff.dev/content/feu2/javascript-2/module-5/algorithms/sorting-and-searching/selection-sort/selection-sort.png)](https://mollify.noroff.dev/content/feu2/javascript-2/module-5/algorithms/sorting-and-searching/selection-sort/selection-sort.png)


````python
def selection_sort(arr) -> None:
  n = len(arr)
  for i in range(n-1): # n elemanlı bir liste için n-1 tur yeterlidir.Örnek : 2 elemanlı bir listede ilk turda küçük eleman bulunup ilk sıraya konur ve liste sıralanmış olur 2. tura gerek yok.
    min_index = i # Her turda ilk indexteki elemanı en küçük eleman kabul et.
    for j in range(i+1,n): # Her turda en küçük eleman başa geldiği için baştaki i kadar elemanı tekrar kontrol etmeye gerek yok.
      if arr[j] < arr[min_index] :
        min_index = j # Eğer daha küçük bir eleman varsa min_index'i güncelle.

    # swap
    arr[i], arr[min_index] = arr[min_index], arr[i]
````

````python
arr = [10,62,34,27,14,18,19,52,43,31,11]
````

````python
# Sıralamadan önce liste
print_arr(arr)
````

````python
# Sıralamadan sonra liste
selection_sort(arr)
print_arr(arr)
````

### Insertion Sort  
- İlk adımda ilk eleman ile ikinci eleman kontrol edilir.  
- İkinci adımda 3. eleman sırasıyla 1. ve 2. elemanla karşılaştırılır ve doğru konuma yerleştirilir eğer küçük değilse devam edilir.  
- Liste sonuna kadar aynı işlemler yapılır.  
- Zaman Karmaşıklığı: O(n²)  

[![](https://upload.wikimedia.org/wikipedia/commons/9/9c/Insertion-sort-example.gif)](https://upload.wikimedia.org/wikipedia/commons/9/9c/Insertion-sort-example.gif)


````python
def insertion_sort(arr) -> None:
  n = len(arr)

  for i in range(1,n): # 1'den başlamanılmasının sebebi ilk elemanın zaten sıralı olduğunun farz edilmesidir.
    key = arr[i] # key, şu anda sıralanmakta olan elemandır. Bu elemanı doğru konuma yerleştireceğiz.
    j = i # j, key'in solundaki elemanları kontrol etmek için kullanılan indeks.

    while j > 0 and key < arr[j-1] :
      arr[j] = arr[j-1] # Solumuzdaki eleman büyükse, onu sağa (bizim boşluğa) kaydırıyoruz.
      j -= 1 # j yi bir öncekine atıyoruz (döngünün devamı için)

    arr[j] = key # döngünün sonunda j'yi 1 azalttıgımız için j+1 artık key'i ekleyeceğimiz doğru yer oluyor.
````

````python
arr = [10,62,34,27,14,18,19,52,43,31,11]
````

````python
# Sıralamadan önce liste
print_arr(arr)
````

````python
# Sıralamadan sonra liste
insertion_sort(arr)
print_arr(arr)
````

### Quick Sort  
- Pivot seçilir: Diziden bir öğe pivot olarak seçilir. Pivot seçimi değişebilir (örneğin, ilk öğe, son öğe, rastgele öğe veya medyan).  
- Dizi bölümlere ayrılır: Dizi pivot etrafında yeniden düzenlenir. Bölümleme işlemi sonucunda pivotdan küçük tüm elemanlar sol tarafa, pivotdan büyük tüm elemanlar ise sağ tarafa yerleştirilir. Bu işlem sonunda pivot doğru konumdadır ve pivotun yeni indeksi elde edilir.  
- Özyinelemeli (rekürsif) çağrı yapılır: Aynı işlemi iki bölümlü alt diziye (pivotun solunda ve sağındaki iki diziye) özyinelemeli olarak uygulanır ve liste sıralanmış olur.    
- Zaman Karmaşıklığı : O(n log n)   
[Sorting_quicksort_anim.gif](https://en.wikipedia.org/wiki/Quicksort#/media/File:Sorting_quicksort_anim.gif)

````python
# Partition fonksiyonu öncelikle bir pivot seçer ve pivotu pivottan küçükler ve pivottan büyükler olmak üzere 2'ye ayırır.
# Pivotu dizide doğru konumuna yerleştirir sol tarafında pivotdan küçük, sağ tarafında pivotdan büyük elemanlar olur.(Tam sıralama gerekmiyor)
# Örnek olarak arr = [9,2,3,7,6,4] olsun pivot 4 seçilirse partition fonksiyonu diziyi su hale getirir : arr=[3,2,4,9,6,7]

def partition(arr,start,end) -> int:
  pivot = arr[end] # pivot olarak son elemanı seçiyoruz.
  i = start-1 # pivottan küçük elemanların son indeksini temsil edecek yani pivotun bir solu
  for j in range(start,end):
    if arr[j] < pivot :
      i += 1
      arr[i], arr[j] = arr[j], arr[i] # pivottan küçük olan elemanları sola tasıyoruz.
  arr[i+1], arr[end] = arr[end], arr[i+1] # Pivotu , küçük elemanların sonuna tasıyoruz. Yani pivotun solu pivottan küçükler sağı ise pivottan büyükler haline geliyor. Daha spnra rekürsif çağrılar ile sağ ve sol tarafı sıralıyoruz.
  return i+1
````

````python
# Rekürsif quick sort
# Nesneler üzerinde işlem yaptgımızdan fonksiyon çağırıldıgında direkt olarak liste değişir.
def quick_sort(arr,start,end) -> None:
  if start < end :
     pivot_index = partition(arr,start,end) # Listeyi pivottan küçük olanlar , pivot ve pivottan büyük olanlar seklinde ayırır ve pivotun indeksini döndürür.
     quick_sort(arr,start,pivot_index-1)
     quick_sort(arr,pivot_index+1,end)
````

````python
arr = [10,62,34,27,14,18,19,52,43,31,11]
````

````python
# Sıralamadan önce liste
print_arr(arr)
````

````python
# Sıralamadan sonra liste
quick_sort(arr,0,len(arr)-1)
print_arr(arr)
````

### Merge Sort  
- Sıralı olmayan diziyi ortadan eşit olarak iki alt diziye ayrılır.  
- Bu ayırma işlemi, alt diziler en çok iki elemanlı olana kadar devam eder.
- Alt diziler kendi içinde sıralanır.Sıralı iki alt diziy tek bir sıralı dizi olacak şekilde birleştirilir.
- Zaman Karmaşıklığı : O(n log n)


[Merge-sort-example-300px.gif](https://en.wikipedia.org/wiki/File:Merge-sort-example-300px.gif#/media/File:Merge-sort-example-300px.gif)



````python
def merge_sort(arr) -> list:
  if len(arr) <= 1 :
    return arr
  middle = len(arr) // 2
  leftside = merge_sort(arr[:middle]) #  Ortanca elemana kadar olan kısım sol kısım, ortanca elemandan sonrası sağ kısım
  rightside = merge_sort(arr[middle:])
  return merge(leftside,rightside)
````

````python
def merge(leftside,rightside) -> list:
  result = []
  i = j = 0 # i sol kısım için, j sağ kısım için
  while i < len(leftside) and j < len(rightside) :
    if leftside[i] < rightside[j]:
      result.append(leftside[i])
      i += 1
    else :
      result.append(rightside[j])
      j += 1

  # Eğer i ve j'den biri kendi listesini bitirirse diğer listede kalan elemanları results'un sonuna eklelemeliyiz
  result.extend(leftside[i:]) # [x:] , [:x] islemleri liste geri döndürür bu yüzden extend ile tek tek ekliyoruz.'append' kullansaydık direkt liste eklenirdi,'extend' bir listedeki elemanları belirtilen listeye tek tek ekler.Bu durumu hem algoritma ile uyuşmadıgından hem tuple aynı tipteki verileri tuttugundan istenmeyen bir durumdur.
  result.extend(rightside[j:])
  return result
````

````python
unsorted_arr = [85,4,32,15.7,785,-89,-52.14,13,0]
print("Unsorted array:", unsorted_arr)
````

````python
sorted_arr = merge_sort(unsorted_arr)
print("Sorted array:", sorted_arr)
````

### Shell Sort  
- İlk olarak dizinin uzunluğuna göre bir gap (aralık) değeri belirlenir.Her turda gap değeri yarıya düşürülür.  
- Gap değerine göre, listedeki elemanlar belirli aralıklarla karşılaştırılır ve insertion sort mantığıyla yerlerine yerleştirilir.  
- Shell Sort, Insertion Sort’un geliştirilmiş, çok daha hızlı bir versiyonudur.
Insertion Sortta büyük bir eleman en baştaysa çok fazla kaydırma gerekiyor.
Shell Sort bu sorunu çözüyor.(İnsertion sort gap'ın 1 oldugu durumdur.Ancak önceki karşılaştırmalar sıralama hızını arttırır.)
- Diziyi önce büyük mesafelerle karşılaştırarak sıralıyor sonra adım adım bu mesafeyi azaltıyor.En sonda normal Insertion Sort çalışıyor ancak dizi daha sıralı bir biçimde oldugundan klasik versiyondan daha hızlı biçimde çalışıyor.
- Zaman Karmaşıklığı: O(n²)

[Sorting_shellsort_anim.gif](https://en.wikipedia.org/wiki/Shellsort#/media/File:Sorting_shellsort_anim.gif)


````python
def shell_sort(arr) -> None:
  n = len(arr)
  gap = n // 2 # Başlangıç aralığı

  # gap 1 olana kadar (yani standart insertion sort yapana kadar) devam et
  while gap > 0 :
    for i in range(gap,n): # range(gap, n) olması ilk gap kadar elemanın zaten kendi alt gruplarının ilk elemanı kabul edilmesindendir.
      key = arr[i] # karşılaştırılacak ve araya sokulacak eleman
      j = i # yer değiştirecek elemanın indeksi

      while j >= gap and key < arr[j-gap] : # j >= gap: Listenin başına taşmamak için kontrol. arr[j-gap] > temp: Eğer "gap kadar gerideki" eleman, bizim elimizdeki (temp) sayıdan büyükse
        arr[j] = arr[j-gap]
        j -= gap

      arr[j] = key
    gap //= 2
````

````python
arr = [12, 34, 54, 2, 3]
# Shell sort algoritması uygulanmamış liste :
print_arr(arr)
````

````python
# Shell sort algoritması uygulanmış liste :
shell_sort(arr)
print_arr(arr)
````

- Bu işlem en büyük basamağa kadar sürer.Sonunda liste sıralanmış olur.

````python
def counting_sort(arr,exp) -> None:
  n = len(arr)
  output = [0] * n # n sayıda elemanlı, 0 değerli liste oluşturur.sıralanmış sayıları geçici olarak tutacağımız dizi.
  count = [0] * 10 # 0 dan 9 a kadar olan basamakları tutmak için bir 0 değerli liste.

  for i in range(n):
    index = (arr[i]//exp)%10 # (arr[i]//exp)%10 : istenilen basamak değeri (birler, onlar vs.)
    count[index] += 1 # Bahsi geçen basamakta her rakamın frekansını tutar 0-9

  # Bu döngü her rakamın listede hangi indexe kadar yer kaplayacağını hesaplar.Örn: count[5] = 3 ise, sonu 5 olan elemanlar en fazla 3. sıraya (index 2) yerleşebilir.
  for i in range(1,10):
    count[i] += count[i-1] # Kümülatif olarak tutmaya baslar

  i = n-1 # Döngüyü sondan başa kurmamızın sebebi sayıların doğru yerleşmesidir.Eğer 45 ve 75 gibi sonu aynı olan iki sayı varsa orjinal listede 75 45'ten sonraysa yeni listede de sonra gelmelidir.Tersten başlamak bunu garanti eder.Yani aynı rakama sahip sayılardan orijinal listede sonda olan yine sonda kalsın.
  while i >= 0:
    index = (arr[i]//exp)%10
    output[count[index]-1] = arr[i] # count[index] bize sayının kaçıncı sırada olduğunu verir.indeks 0'dan başladığı için 1 çıkarıp doğru yere koyuyoruz.
    count[index] -= 1 # Aynı rakamdan bir tane daha gelirse bu yerleştiğimiz yerin bir soluna yerleşsin diye sayacı azaltıyoruz.
    i -= 1

  i=0
  for i in range(0,n):
    arr[i] = output[i] # Sıralanmış geçici listeyi asıl listemize aktarıyoruz.
````

````python
def radix_sort(arr) -> None:
  max1 = max(arr) # Find the maximum element in the array
  exp = 1
  while max1 / exp >= 1:
    counting_sort(arr, exp)
    exp *= 10
````

````python
arr = [170, 45, 75, 90, 802, 24, 2, 66]
# Radix sort algoritması uygulanmamış liste :
print_arr(arr)
````

````python
# Radix sort algoritması uygulanmış liste :
radix_sort(arr)
print_arr(arr)
````

### Heap Sort  
- ***heap*** : Her düğümün belirli bir sıraya (max veya min) göre düzenlendiği tamamlanmış ikili ağaç (binary tree).
- ***max heap*** her parent düğümün çocuklarından **büyük** olması anlamına gelir.***min heap*** ise her parent düğümün çocuklarından **küçük** olması anlamına gelir.
Örnek :

[![](https://images.wondershare.com/edrawmax/articles2024/max-heap-python-guide/how-to-build-max-heap-data-structure-02.png)](https://images.wondershare.com/edrawmax/articles2024/max-heap-python-guide/how-to-build-max-heap-data-structure-02.png)   
- ***built-max-heap*** : sıralanmamış olan listeyi 'max heap' haline getirme işlemidir.  
- ***heapify*** : ***buil-max-heap***'e benzerlik gösterir ama ***heapify*** listenin bir kısmının zaten sıralanmış olduğunu farzeder.  
- Heap Sort Adımları:
  - İlk olarak liste ***max heap*** haline getirilir.Daha sonra listenin en büyük elemanı bulunur ve listenin sonuna taşınır.Sonrasında ***heap*** 1 azaltılır ve ***heapify*** fonksiyonu ile liste yeniden ***max heap*** haline getirilir.Tüm elemanlar sıralanana kadar bu işlemler tekrarlanır.

- İkili ağacı dizide göstermek :
  - i → kök
  - 2*i+1 → sol çocuk
  - 2*i+2 → sağ çocuk  


  [![](https://media.geeksforgeeks.org/wp-content/uploads/20241007174934849696/Array-Representation-of-Binary-Tree.webp)](https://media.geeksforgeeks.org/wp-content/uploads/20241007174934849696/Array-Representation-of-Binary-Tree.webp)

  - Son katmanda (yani leaf'lerin(çocuğu olmayan düğümlerin) bulunduğu katmanda) indeks n//2'den başlar ve n-1'de biter. Yani son yaprak olmayan düğümün indeksi n//2- 1'dir.

````python
'''
Verilen düğümü kök kabul eder ve onun altındaki çocukları kontrol eder.Eğer sol veya sağ çocuk kökten büyükse, kökle yer değiştirir.Ama bu değişimden sonra yeni kökün altında daha büyük bir eleman olabilir
işte bu yüzden rekürsif çağrı yapılır.
'''
def heapify(arr, n, i):
    largest = i          # En büyük elemanı kök olarak varsay
    left = 2 * i + 1     # Sol çocuk
    right = 2 * i + 2    # Sağ çocuk

    # Dizinin boyutu aşılmamışsa ve sol çocuk kökten büyükse
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Dizinin boyutu aşılmamışsa ve sağ çocuk kökten büyükse
    if right < n and arr[right] > arr[largest]:
        largest = right


    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)  # Alt ağaçta tekrar heapify
````

````python
def heap_sort(arr):
    n = len(arr)

    # İlk for diziyi max heap'e dönüştürmek için var. Leaf düğümleri heapify etmiyoruz çünkü zaten çocukları yok.
    # n//2 - 1 leaf olmayan son düğümdür; buradan başlayıp root'a kadar gidiyoruz.
    # Sondan başlamamızın sebebi, önce alt ağaçları düzenlememiz gerektiği.
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i) # Hep aynı ağaç üzerinde çalışıyoruz; i hangi düğümden heapify'e başlanacağını gösteriyor.

    for i in range(n-1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]  # heapify işlemlerinden sonra en büyük eleman kökte (yani dizinin ilk indeksi) olur. En büyük elemanı sona taşıyoruz.

        # Burada sadece kök bozulur. O yüzden yeniden tüm düğümleri heapify etmeye gerek yok (yukarıdaki ilk for döngüsü gibi).
        # Kalan kısmı yeniden max heap yapmak için yalnızca kökten başlayarak heapify yapmak yeterli.
        heapify(arr, i, 0)
````

````python
arr = [12, 11, 13, 5, 6, 7]
# Heap sort algoritması uygulanmamış liste :
print_arr(arr)
````

````python
heap_sort(arr)
print_arr(arr)
````
